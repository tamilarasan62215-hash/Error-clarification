# selection k algoritham

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
def select_kbest(indep_x,dep_y,n):
    test=SelectKBest(score_func=chi2,k=n)
    fit1=test.fit(indep_x,dep_y)
    select_kfeatures=fit1.transform(indep_x)
    Scolumns=test.get_feature_names_out()
    print(Scolumns)
    #selected_colum=indep_x.columns[test.get_support()]
    #print(selected_colum)
    return select_kfeatures
```

```python
indep_x=dataset1.drop("LUNG_CANCER_YES",axis=1)
dep_y=dataset1["LUNG_CANCER_YES"]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
def split_scaler(indep_x,dep_y):
    x_train,x_test,y_train,y_test=train_test_split(indep_x,dep_y,test_size=0.25,random_state=0)
    scx=StandardScaler()
    x_train=scx.fit_transform(x_train)
    x_test=scx.transform(x_test)
    return x_train,x_test,y_train,y_test
```

```python
def cm_prediction(classifier,x_test):
    y_pred=classifier.predict(x_test)

    from sklearn.metrics import confusion_matrix
    cm=confusion_matrix(y_test,y_pred)

    from sklearn.metrics import accuracy_score
    accuracy=accuracy_score(y_test,y_pred)

    from sklearn.metrics import classification_report
    cl_report=classification_report(y_test,y_pred)

    return classifier,accuracy,cl_report,x_test,y_test,cm
```

```python
from sklearn.linear_model import LogisticRegression
def logistic(x_train,y_train,x_test):
    classifier=LogisticRegression()
    classifier.fit(x_train,y_train)
    classifier,accuracy,cl_report,x_train,y_train,cm=cm_prediction(classifier,x_test)
    return classifier,accuracy,cl_report,x_train,y_train,cm
```

```python
from sklearn.svm import LinearSVC
def svm_linear(x_train,y_train,x_test):
    classifier=LinearSVC()
    from sklearn.svm import SVC
    classifier.fit(x_train, y_train)
    classifier,Accuracy,report,x_test,y_test,cm=cm_prediction(classifier,x_test)
    return  classifier,Accuracy,report,x_test,y_test,cm    
```

```python
from sklearn.svm import SVC
def svm_NL(x_train,y_train,x_test):
    classifier=SVC()
    classifier.fit(x_train, y_train)
    classifier,Accuracy,report,x_test,y_test,cm=cm_prediction(classifier,x_test)
    return  classifier,Accuracy,report,x_test,y_test,cm    
```

```python
from sklearn.neighbors import KNeighborsClassifier
def knn(x_train,y_train,x_test):
    classifier=KNeighborsClassifier()
    classifier.fit(x_train, y_train)
    classifier,Accuracy,report,x_test,y_test,cm=cm_prediction(classifier,x_test)
    return  classifier,Accuracy,report,x_test,y_test,cm    
```

```python
from sklearn.tree import DecisionTreeClassifier
def Decision(x_train,y_train,x_test):
    classifier=DecisionTreeClassifier()
    classifier.fit(x_train, y_train)
    classifier,Accuracy,report,x_test,y_test,cm=cm_prediction(classifier,x_test)
    return  classifier,Accuracy,report,x_test,y_test,cm
```

```python
from sklearn.ensemble import RandomForestClassifier
def random(x_train,y_train,x_test):
    classifier=RandomForestClassifier()
    classifier.fit(x_train, y_train)
    classifier,Accuracy,report,x_test,y_test,cm=cm_prediction(classifier,x_test)
    return  classifier,Accuracy,report,x_test,y_test,cm
```

```python
def selectk_classification(acc_log_r,acc_svm_l,acc_svm_nl,acc_knn,acc_dec,acc_rf):
    dataframe=pd.DataFrame(index=["chisquare"],columns=["Logistic","svml","svmnl","knn","Decision","random"])
    for number,idex in enumerate(dataframe.index):
        dataframe["Logistic"][idex]=acc_log_r[number]
        dataframe["svml"][idex]=acc_svm_l[number]
        dataframe["svmnl"][idex]=acc_svm_nl[number]
        dataframe["knn"][idex]=acc_knn[number]
        dataframe["Decision"][idex]=acc_dec[number]
        dataframe["random"][idex]=acc_rf[number]
        return dataframe
```

```python
acc_log_r=[]
acc_svm_l=[]
acc_svm_nl=[]
acc_knn=[]
acc_dec=[]
acc_rf=[]
```

```python
kbest=select_kbest(indep_x,dep_y,7)
kbest
```

```python
x_train,x_test,y_train,y_test=split_scaler(kbest,dep_y)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)

classifier,Accuracy,report,x_test,y_test,cm=logistic(x_train,y_train,x_test)
acc_log_r.append(Accuracy)

classifier,Accuracy,report,x_test,y_test,cm=svm_linear(x_train,y_train,x_test)
acc_svm_l.append(Accuracy)

classifier,Accuracy,report,x_test,y_test,cm=svm_NL(x_train,y_train,x_test)
acc_svm_nl.append(Accuracy)

classifier,Accuracy,report,x_test,y_test,cm=knn(x_train,y_train,x_test)
acc_knn.append(Accuracy)

classifier,Accuracy,report,x_test,y_test,cm=Decision(x_train,y_train,x_test)
acc_dec.append(Accuracy)

classifier,Accuracy,report,x_test,y_test,cm=random(x_train,y_train,x_test)
acc_rf.append(Accuracy)


result=selectk_classification(acc_log_r,acc_svm_l,acc_svm_nl,acc_knn,acc_dec,acc_rf)
```

# RFE(recursive feature elimination)

```python
from sklearn.feature_selection import RFE
# finde best algorithem with sutable columns

def rfeFeature (indep_x,dep_y,n): # select the best column with reject same evaluation valu
    rfelist=[]
    log_model=LogisticRegression()#solver="lbfgs"
    SVC_L=LinearSVC()#random_state=0, tol=1e-5
    SVC_NL=SVC()
    KNN= KNeighborsClassifier()#n_neighbors=3
    RF=RandomForestClassifier()
    DT=DecisionTreeClassifier()#criterion="gini",max_features="aqrt",splitter="best",random_state=0
    RFEmodel_list=[log_model,SVC_L,SVC_NL,KNN,RF,DT]
    for i in RFEmodel_list:
        print(i)
        log_rfe=RFE(estimator=i,n_features_to_select=n)
        log_fit=log_rfe.fit(indep_x,dep_y)
        log_rfe_feature=log_fit.transform(indep_x)
        rfelist.append(log_rfe_feature)
        return rfelist
    selected_colum=indep_x.columns[log_fit.get_support()]
    print(selected_colum)
```

```python
rfelist=rfeFeature(indep_x,dep_y,6)
```

```python
acclog=[]
accsvml=[]
accsvmnl=[]
accknn=[]
accdes=[]
accrf=[]
```

```python
def rfe_classification(acclog,accsvml,accsvmnl,accknn,accdes,accrf): 
    
    rfedataframe=pd.DataFrame(index=['Logistic','SVM_L','SVC_NL','KNN','Decision_Tree','Random'],columns=['Logistic','SVMl','SVMnl','KNN','Decision','Random',])
    for number,idex in enumerate(rfedataframe.index):
        
        rfedataframe['Logistic'][idex]=acclog[number]       
        rfedataframe['SVMl'][idex]=accsvml[number]
        rfedataframe['SVMnl'][idex]=accsvmnl[number]
        rfedataframe['KNN'][idex]=accknn[number]
        rfedataframe['Decision'][idex]=accdes[number]
        rfedataframe['Random'][idex]=accrf[number]
    return rfedataframe
```

```python
for i in rfelist:   
    x_train, x_test, y_train, y_test=split_scaler(i,dep_y)   
    
        
    classifier,Accuracy,report,x_test,y_test,cm=logistic(x_train,y_train,x_test)
    acclog.append(Accuracy)
    
    classifier,Accuracy,report,x_test,y_test,cm=svm_linear(x_train,y_train,x_test)  
    accsvml.append(Accuracy)
    
    classifier,Accuracy,report,x_test,y_test,cm=svm_NL(x_train,y_train,x_test)  
    accsvmnl.append(Accuracy)
    
    classifier,Accuracy,report,x_test,y_test,cm=knn(x_train,y_train,x_test)  
    accknn.append(Accuracy)
    
    classifier,Accuracy,report,x_test,y_test,cm=Decision(x_train,y_train,x_test)  
    accdes.append(Accuracy)
    
    classifier,Accuracy,report,x_test,y_test,cm=random(x_train,y_train,x_test)  
    accrf.append(Accuracy)
    
result=rfe_classification(acclog,accsvml,accsvmnl,accknn,accdes,accrf)
```
