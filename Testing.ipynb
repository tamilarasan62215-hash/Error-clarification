```python
class  UnivariteTC():
    descriptive=pd.DataFrame(index=["mean","median","mode","Q1:25%","Q2:50%","Q3:75%","99%","Q4:100%","IQR","1.5rule","lesser","greater","min","max"],columns=quan)
for columnName in quan:
    descriptive[columnName]["mean"]=dataset[columnName].mean()
    descriptive[columnName]["median"]=dataset[columnName].median()
    descriptive[columnName]["mode"]=dataset[columnName].mode()[0]
    descriptive[columnName]["Q1:25%"]=dataset.describe()[columnName]["25%"]
    descriptive[columnName]["Q2:50%"]=dataset.describe()[columnName]["50%"]
    descriptive[columnName]["Q3:75%"]=dataset.describe()[columnName]["75%"]
    descriptive[columnName]["99%"]=np.percentile(dataset[columnName],99)       # 99% salary valu not in describe so using percentile 
    descriptive[columnName]["Q4:100%"]=dataset.describe()[columnName]["max"]      # in ["max"] column
    descriptive[columnName]["IQR"]=descriptive[columnName]["Q3:75%"]-descriptive[columnName]["Q1:25%"]   #(Q3 percentile - Q1 percentile)
    descriptive[columnName]["1.5rule"]=1.5*descriptive[columnName]["IQR"]  #(1.5 * IQR value)
    descriptive[columnName]["lesser"]=descriptive[columnName]["Q1:25%"]-descriptive[columnName]["1.5rule"]  #(Q1-1.5rule)
    descriptive[columnName]["greater"]=descriptive[columnName]["Q3:75%"]+descriptive[columnName]["1.5rule"] #(Q3+1.5rule)
    descriptive[columnName]["min"]=dataset[columnName].min() # dataset minimum value
    descriptive[columnName]["max"]=dataset[columnName].max() # dataset maxmium value
return descriptive
```

```python
def freqTable(columnName,dataset):
    freqTable=pd.DataFrame(columns=["Unique_values","Frequency","Relative_Frequency","cumsum"])
    freqTable["Unique_values"]=dataset[columnName].value_counts().index
    freqTable["Frequency"]=dataset[columnName].value_counts().values
    freqTable["Relative_Frequency"]=(freqTable["Frequency"]/103)
    freqTable["cumsum"]=freqTable["Relative_Frequency"].cumsum()
    return freqTable
```
