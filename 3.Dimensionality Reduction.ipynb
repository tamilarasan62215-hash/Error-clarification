```python
def cm_prediction(classifier,x_test):  # confus matrix (evlovation matrix for classification model)
    y_pred=classifier.predict(x_test)  # (predict_value,x_test)
         
    from sklearn.metrics import confusion_matrix
    cm=confusion_matrix(y_test,y_pred)
       
    from sklearn.metrics import accuracy_score
    from sklearn.metrics import classification_report

    Accuracy=accuracy_score(y_test,y_pred)
    
    report =classification_report(y_test,y_pred)                    

    return classifier,Accuracy,report,x_test,y_test,cm
```

```python
def logistic(x_train,y_train):
    
    from sklearn.linear_model import LogisticRegression
    classifier=LogisticRegression(random_state=0)
    classifier.fit(x_train,y_train)
    classifier,Accuracy,report,x_test,y_test,cm=cm_prediction(classifier,x_test)
    return classifier,Accuracy,report,x_test,y_test,cm
```

```python
def svm_linear(x_train,y_train):
    from sklearn.svm import SVC
    classifier=SVC(kernel="linear",random_state=0)
    classifier.fit(x_train,y_train)
    classifier,Accuracy,report,x_train,y_train,cm=cm_prediction(classifier,x_test)
    return classifier,Accuracy,report,x_test,y_test,cm
```

```python
def svm_NL(x_train,y_train):
    from sklearn.svm import SVC
    classifier=SVC(kernel="rbf",random_state=0)
    classifier.fit(x_train,y_train)
    classifier,Accuracy,report,x_train,y_train,cm=cm_prediction(classifier,x_test)
    return classifier,Accuracy,report,x_test,y_test,cm
```

```python
def Navie(x_train,y_train):
    from sklearn.naive_bayes import GaussianNB
    classifier=GaussianNB()
    classifier.fit(x_train,y_train)
    classifier,Accuracy,report,x_train,y_train,cm=cm_prediction(classifier,x_test)
    return classifier,Accuracy,report,x_test,y_test,cm
```

```python
def knn(x_train,y_train):
    from sklearn.neighbors import KNeighborsClassifier
    classifier=KNeighborsClassifier(n_neighbors=5,metric="minkowski",p=2)
    classifier.fit(x_train,y_train)
    classifier,Accuracy,report,x_train,y_train,cm=cm_prediction(classifier,x_test)
    return classifier,Accuracy,report,x_test,y_test,cm                               
```

```python
def Decision(x_train,y_train):
    from sklearn.tree import DecisionTreeClassifier
    classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
    classifier.fit(x_train, y_train)  
    classifier,Accuracy,report,x_test,y_test,cm=cm_prediction(classifier,x_test)
    return  classifier,Accuracy,report,x_test,y_test,cm
```

```python
def random(x_train,y_train):
        from sklearn.ensemble import RandomForestClassifier
        classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
        classifier.fit(x_train, y_train)
        classifier,Accuracy,report,x_test,y_test,cm=cm_prediction(classifier,x_test)
        return  classifier,Accuracy,report,x_test,y_test,cm
```

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 1. Create a dummy dataset (100 samples, 5 features)
np.random.seed(42)
data = np.random.rand(100, 5)
columns = ['Feature_1', 'Feature_2', 'Feature_3', 'Feature_4', 'Feature_5']
df = pd.DataFrame(data, columns=columns)

# 2. Standardize features (Crucial step for PCA)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# 3. Initialize PCA (Specify how many components you want)
# You can pass an integer (e.g., 2) or a float for variance (e.g., 0.95)
pca = PCA(n_components=2)

# 4. Fit and transform the data
pca_result = pca.fit_transform(scaled_data)

# 5. Convert results to a readable DataFrame
df_pca = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])

# --- Key Metrics Output ---
print("--- PCA Summary ---")
print(f"Original Shape: {scaled_data.shape}")
print(f"Reduced Shape: {df_pca.shape}\n")

# Explained Variance Ratio shows how much information each component holds
print("Variance explained by each component:")
for i, ratio in enumerate(pca.explained_variance_ratio_):
    print(f"  Principal Component {i+1}: {ratio:.2%}")

print(f"\nTotal variance retained: {np.sum(pca.explained_variance_ratio_):.2%}")
```
